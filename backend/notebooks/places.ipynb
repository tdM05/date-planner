```python
import os
import json
import googlemaps
from dotenv import load_dotenv
from pathlib import Path

# Load environment variables from your .env file in the root directory
env_path = Path('..') / '.env' # Assumes notebook is in a subfolder
load_dotenv(dotenv_path=env_path)

GOOGLE_MAPS_API_KEY = os.getenv("GOOGLE_MAPS_API_KEY")

# Add a check to ensure the key was loaded
if not GOOGLE_MAPS_API_KEY:
    print("Error: GOOGLE_MAPS_API_KEY not found. Check your .env file and path.")
else:
    print("Google Maps API Key loaded successfully.")
    gmaps_client = googlemaps.Client(key=GOOGLE_MAPS_API_KEY)
```

```python
# This is a mock response, simulating the output from your first Claude API call
claude_suggestion = {
    "activity_type": "Restaurant",
    "location_query": "Cozy romantic Italian restaurant in Little Italy, Toronto"
}

print(f"Starting with the suggestion: '{claude_suggestion['location_query']}'")
```

```python
place_id = None

try:
    print("\n--- Calling Google Places API: Text Search ---")
    # Use the 'places' method which corresponds to a Text Search
    search_results = gmaps_client.places(query=claude_suggestion['location_query'])
    
    # Check if we got any results
    if search_results['status'] == 'OK' and search_results['results']:
        # Get the top result
        top_result = search_results['results'][0]
        place_id = top_result['place_id']
        place_name = top_result['name']
        print(f"Found a match: '{place_name}' with place_id: {place_id}")
    else:
        print(f"Could not find any results for the query. Status: {search_results.get('status')}")

except Exception as e:
    print(f"An error occurred during Text Search: {e}")
```
