```python
import requests
import os
from dotenv import load_dotenv

# Make sure you've loaded your .env file
# load_dotenv(...) 

OPENWEATHER_API_KEY = os.getenv("OPENWEATHER_API_KEY")
print(OPENWEATHER_API_KEY)
```

```python

# Coordinates for Toronto (you would get these from the Google Places API)
lat = 43.6532
lon = -79.3832

# Construct the API request URL for the One Call API
# The 'exclude' part is optional, it just cleans up the response
# CORRECTED URL: Use the 2.5/weather endpoint for current conditions
url = f"https://api.openweathermap.org/data/2.5/forecast?lat={lat}&lon={lon}&appid={OPENWEATHER_API_KEY}"

try:
    response = requests.get(url)
except requests.exceptions.HTTPError as err:
    print(f"HTTP Error: {err}")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
response_json = response.json()
print(response_json)
```
